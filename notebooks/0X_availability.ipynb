```python
import pandas as pd
import os
```

```python
path_trips=os.path.join(os.getcwd(), "..", "data", "output", "Trips.csv")

print(path_trips)
```

```python
df_trips = pd.read_csv(path_trips, index_col="Unnamed: 0")
df_trips
```

```python
df_trips["datetime_start"] = pd.to_datetime(df_trips["datetime_start"])
df_trips["datetime_end"] = pd.to_datetime(df_trips["datetime_end"])
df_trips["24_agg_time"] = pd.to_datetime(df_trips["24_agg_time"])
df_trips["1_agg_time"] = pd.to_datetime(df_trips["1_agg_time"])
df_trips.columns
```

```python
# DELETE ME
df_trips = df_trips[["datetime_start", "datetime_end", "h3_hex_big_id_start", "h3_hex_big_id_end", "24_agg_time", "24_agg_time_end", "24_demand_hex_big", "24_demand_hex_big_end", "b_number_start"]].copy()
```

```python
# df_trips["24_available_hex_big"] = df_trips["24_demand_hex_big_end"] - df_trips["24_demand_hex_big"]
# df_trips[df_trips["24_available_hex_big"] != 0]
```

```python
df_current_hex = df_trips[df_trips["h3_hex_big_id_end"] == "881f1b4601fffff"]
df_current_hex.head(100)
```

```python
df_end_before_agg_time = df_current_hex[df_current_hex["datetime_end"] < "2019-09-20"]
df_end_before_agg_time
```

```python
bikes_before = df_end_before_agg_time.groupby("b_number_start")["datetime_end"].max().count()
bikes_before
```

```python
df_start_before_agg_time_end = df_current_hex[df_current_hex["datetime_start"] > "2019-09-21"]
df_start_before_agg_time_end
```

```python
bikes_used_in_agg_time = df_start_before_agg_time_end.groupby("b_number_start")["datetime_start"].count().count()
bikes_used_in_agg_time
```

```python
idle_bikes = bikes_before - bikes_used_in_agg_time
idle_bikes
```

```python
df_test = df_trips[(df_trips["h3_hex_big_id_end"] == "881f1b4601fffff") & (df_trips["24_agg_time"] == "2019-09-20")]
df_test
```

```python
df_test["24_idle_bikes_hex_big"] = idle_bikes
df_test["24_available_hex_big"] = df_test["24_demand_hex_big_end"] - df_test["24_demand_hex_big"] + idle_bikes
df_test.tail(20)
```
