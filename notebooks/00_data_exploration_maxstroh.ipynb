```python
import os
import pandas as pd
import glob
```

```python
path=os.path.join(os.getcwd(), '../data/input/dresden/2019/')
path
```

```python
directors = glob.glob(path+"*")
directors

all_files = []
for director in directors:
    all_files=all_files+glob.glob(director+"/*.csv")
```

```python
li = []

for filename in all_files:
    df_read = pd.read_csv(filename, index_col=None, header=0)
    li.append(df_read)

df = pd.concat(li, axis=0, ignore_index=True)
```

# Dropping unused columns
## Columns including single value
The following columns only included one value (or NaN values)
* city : ['dresden']
* p_rack_locks :  [False]
* b_state :  ['ok']
* b_active :  [ True]
* b_battery_pack :  [nan '{"percentage": 0}']
* p_special_racks :  [nan  0.]
* p_free_special_racks :  [nan  0.]

## Columns which highly correlate
The following columns highly correlate - meaning they are complementary booleans or if one is high, the other is too
* p_bike [corr(p_spot)=-0.999501, complementary boolean to p_spot]

```python
df_cleaned = df.drop(["city", "p_bike", "p_rack_locks", "b_active", "b_state", "b_battery_pack", "p_rack_locks", "p_special_racks", "p_free_special_racks"], axis=1)
df_cleaned
```

```python
corrs_cleaned = df_cleaned.corr()
corrs_cleaned
```

```python
corrs_cleaned.ge(0.9) | corrs_cleaned.le(-0.9)
```

## Still highly correlating columns
* p_spot <-> p_place_type (-0.994030)
* b_boardcomputer <-> b_electric_lock (0.993479)
